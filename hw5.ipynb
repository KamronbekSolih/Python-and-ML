```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("data_2d.csv")

print(data.columns.tolist())

x1 = data['x1'].values.reshape(-1,1)
y = data['y'].values.reshape(-1,1)

# Adding one coulmns

x1_ones = np.ones((x1.shape[0],1))

X1 = np.concatenate((x1_ones,x1),axis=1)


W = np.linalg.inv(X1.T@X1)@X1.T@y

test_data = pd.read_csv('test_data_2d.csv')
test_x = test_data['x1'].values.reshape(-1,1)
test_y = test_data['y'].values.reshape(-1,1)

# Add a column of ones to test_x

x_ones_test = np.ones((test_x.shape[0],1))
X_test = np.concatenate((x_ones_test,test_x),axis=1)

pred = X_test@W
MSE = np.mean(np.square(pred-test_y))
print("MSE: ", MSE)

# print(f"y: {y}, y_pred: {pred} ")

# Visualization.

plt.scatter(x1, y, color = 'red', label='Training data')
plt.scatter(test_x, test_y, color = 'black', label='Testing data')  # Corrected from 'test_data'
plt.plot(test_x, pred, 'r', label='Model predictions')  # Corrected from 'test_data'
plt.xlabel("x1")
plt.ylabel("y")
plt.legend()
plt.show()

# with this somehow calculations had been performed not so accurately. As MSE depics the deffirence between actual and predicted values is huge.I leave this model as an exaple how the models should NOT work
```

```python
# Read 'data_1D.cvs' and 'test_data_1D.cvs'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



data =  pd.read_csv('test_data_2d.csv')
x_data = data['x1'].values
y_data = data['y'].values

test_data = pd.read_csv('data_2d.csv')
x_test = test_data['x1'].values
y_test = test_data['y'].values

sum_x = sum(x_data)
sum_y = sum(y_data)
sum_x_sq = sum([x**2 for x in x_data])
sum_xy = []
n = len(x_data)

for i in range(0,n):
    sum_xy.append(x_data[i]*y_data[i])

sum_xy = sum(sum_xy)

m = (n*sum_xy-sum_x*sum_y)/(n*sum_x_sq-sum_x**2)         # slope

b = (sum_y-m*sum_x)/n   # intercept

def y(x):
    return m*x+b

new_y = [ y(x) for x in x_data]
mse = []
for i in range(0,n):
    mse.append(y_data[i]-new_y[i])

mse = sum([a**2 for a in mse])/n

print("MSE: ",mse)


plt.plot(x_data,y(x_data),color='red', marker = 'o', label = 'Model predictions')
plt.scatter(x_test,y_test,label = 'Training data', marker='o')
plt.scatter(x_data,y_data,label = 'Test data' , marker='o', color = 'black')

plt.grid(True)
plt.xlabel("x1")
plt.ylabel('y')
plt.title('Week 4 data sample')
plt.legend()
plt.show()

```

regression result as in Fig. 1 to 3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('test_data_2d.csv')
actual_data = pd.read_csv('data_2d.csv')


x = data[data.columns.tolist()[0]]
y = data[data.columns.tolist()[1]]

x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)

andoza = LinearRegression()

andoza.fit(x,y)

r2 = andoza.score(x,y)
b0 = andoza.intercept_
b1 = andoza.coef_

y_pred = b0 +b1*x
print(f"x = 60 y = {b0+b1*60}")
print(f"r2 = {r2} \nb0 = {b0} \nb1 = {b1} \ny_pred = {y_pred[:10]}")

# Visualizing the data

plt.scatter(x,y, label = 'Actual values')
plt.scatter()
plt.plot(x,y_pred, label = 'Predicted values', color = 'r')
plt.xlabel("y")
plt.ylabel("x1")
plt.legend()
plt.show()
```
